# Linear Search (No Sorting Required)/Búsqueda lineal (no requiere clasificación)/:

- Time Complexity: $O(n)$

Español:

- Complejidad temporal: $O(n)$


```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# Example usage:
arr = [3, 6, 1, 8, 10, 2, 1]
target = 6

linear_search_result = linear_search(arr, target)
print("Original array:", arr)
print("Linear Search Result:", linear_search_result)
```

# Interpolation Search (Requires Sorting with Quicksort)/Búsqueda por interpolación (requiere ordenación con Quicksort):

- The time complexity of Interpolation Search, when used with a sorted array (achieved through Quicksort or any other sorting algorithm), is typically considered to be $O(\log \log n)$ on average, where $n$ is the number of elements in the array. However, in the worst-case scenario, it can be as bad as $O(n)$.

Español:

- La complejidad temporal de la b￺squeda por interpolaci￳n, cuando se utiliza con una matriz ordenada (conseguida mediante Quicksort o cualquier otro algoritmo de ordenaci￳n), suele considerarse $O(\log \log n)$ de media, donde $n$ es el n￺mero de elementos de la matriz. Sin embargo, en el peor de los casos, puede ser tan malo como $O(n)$.

## Pseudocode/Pseudocódigo:

image.png


```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quicksort(left) + middle + quicksort(right)

def interpolation_search(arr, target):
    sorted_arr = quicksort(arr)
    left, right = 0, len(sorted_arr) - 1
    while left <= right and sorted_arr[left] <= target <= sorted_arr[right]:
        if left == right:
            if sorted_arr[left] == target:
                return left
            return -1
        pos = left + ((right - left) // (sorted_arr[right] - sorted_arr[left])) * (target - sorted_arr[left])
        if sorted_arr[pos] == target:
            return pos
        elif sorted_arr[pos] < target:
            left = pos + 1
        else:
            right = pos - 1
    return -1

# Example usage:
arr = [3, 6, 1, 8, 10, 2, 1]
target = 10

interpolation_search_result = interpolation_search(arr, target)
sorted_arr = quicksort(arr)

print("Original array:", arr)
print("Sorted array:", sorted_arr)
print("Interpolation Search Result:", interpolation_search_result)
```

# Binary Search (Requires Sorting with Quicksort):

- Time complexity: $O(log n)$

Español:

- Complejidad Temporal: $O(log n)$


## Pseudocode/Pseudocódigo:

image.png

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quicksort(left) + middle + quicksort(right)

def binary_search(arr, target):
    sorted_arr = quicksort(arr)
    left, right = 0, len(sorted_arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if sorted_arr[mid] == target:
            return mid
        elif sorted_arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Example usage:
arr = [3, 6, 1, 8, 10, 2, 1]
target = 6

binary_search_result = binary_search(arr, target)
sorted_arr = quicksort(arr)

print("Original array:", arr)
print("Sorted array:", sorted_arr)
print("Binary Search Result:", binary_search_result)
```

# Timsort (NOT THE OPTIMIZED VERSION)/(NO LA VERSIÓN OPTIMIZADA)

- Timsort: Is a hybrid sorting algorithm that uses a combination of insertion sort
and merge sort to efficiently sort arrays of any size. It has an average time
complexity of $O(n log n)$

Español:

- Timsort: Es un algoritmo de ordenación híbrido que utiliza una combinación de ordenación por inserción
y la ordenación por fusión para ordenar eficazmente matrices de cualquier tamaño. Tiene una complejidad
de $O(n log n)$.

## Pseudocode/Pseudocódigo:

image.png

```python
def insertion_sort(arr, left=0, right=None):
    if right is None:
        right = len(arr) - 1

    for i in range(left + 1, right + 1):
        key_item = arr[i]
        j = i - 1
        while j >= left and arr[j] > key_item:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key_item

def merge(arr, left, middle, right):
    left_copy = arr[left:middle + 1]
    right_copy = arr[middle + 1:right + 1]

    left_copy_index = 0
    right_copy_index = 0
    sorted_index = left

    while left_copy_index < len(left_copy) and right_copy_index < len(right_copy):
        if left_copy[left_copy_index] <= right_copy[right_copy_index]:
            arr[sorted_index] = left_copy[left_copy_index]
            left_copy_index += 1
        else:
            arr[sorted_index] = right_copy[right_copy_index]
            right_copy_index += 1
        sorted_index += 1

    while left_copy_index < len(left_copy):
        arr[sorted_index] = left_copy[left_copy_index]
        left_copy_index += 1
        sorted_index += 1

    while right_copy_index < len(right_copy):
        arr[sorted_index] = right_copy[right_copy_index]
        right_copy_index += 1
        sorted_index += 1

def timsort(arr):
    min_run = 32
    n = len(arr)

    for i in range(0, n, min_run):
        insertion_sort(arr, i, min((i + min_run - 1), (n - 1)))

    size = min_run
    while size < n:
        for start in range(0, n, size * 2):
            midpoint = min((start + size - 1), (n - 1))
            end = min((start + size * 2 - 1), (n - 1))

            if midpoint < end:
                merge(arr, start, midpoint, end)

        size *= 2

# Usage:
arr = [64, 25, 12, 22, 11]
timsort(arr)
print("Timsort Sorted array:", arr)
```

# Heap Sort

- Heap Sort: A sorting algorithm that uses a data structure called a heap to
perform the sorting. It has a running time of $O(n log n)$.

Español:

- Ordenación por montón: Un algoritmo de ordenación que utiliza una estructura de datos llamada montón para
realizar la ordenación. Tiene un tiempo de ejecución de $O(n log n)$.


## Pseudocode/Pseudocódigo:

## Pseudocode (Build Max-Heap)/Pseudocódigo (Build Max-Heap):

```python
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

# Usage:
arr = [64, 25, 12, 22, 11]
heap_sort(arr)
print("Heap Sort Sorted array:", arr)
```

# Selection Sort

- Selection Sort: A simple sorting algorithm that works by finding the minimum
element and placing it at the beginning of the list, then repeating the process for
each of the remaining elements. It has a running time of $O(n^2)$ in the worst
case.

Español:

- Ordenación por selección: Un algoritmo de ordenación simple que funciona mediante la búsqueda de la mínima
elemento y colocándolo al principio de la lista, repitiendo el proceso para
cada uno de los elementos restantes. Tiene un tiempo de ejecución de $O(n^2)$ en el peor caso.
en el peor de los casos.

## Pseudocode/Pseudocódigo:

```python
def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

# Usage:
arr = [64, 25, 12, 22, 11]
selection_sort(arr)
print("Selection Sorted array:", arr)
```

# Merge Sort

- Merge Sort: A divide-and-conquer based sorting algorithm that works by dividing
the data into small pieces, sorting them, and then recombining them. It is stable
and has a running time of $O(n log n)$.

Español:

- Ordenación por fusión: Un algoritmo de ordenación basado en dividir y conquistar que funciona dividiendo
los datos en trozos pequeños, los ordena y luego los recombina. Es estable
y tiene un tiempo de ejecución de $O(n log n)$.

## Pseudocode/Pseudocódigo:

image.png

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

# Usage:
arr = [64, 25, 12, 22, 11]
merge_sort(arr)
print("Merge Sorted array:", arr)
```

# Quick Sort

- Quick Sort: A divide-and-conquer sorting algorithm that works by choosing an
element as a pivot and partitioning the data around it. It is one of the fastest
algorithms on average.


Español:

- Ordenación rápida: Un algoritmo de ordenación de divide y vencerás que funciona eligiendo un
elemento como pivote y dividiendo los datos en torno a él. Es uno de los algoritmos
más rápidos.

## Pseudocode/Pseudocódigo:

image.png

## Pseudocode (partitioning):

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Usage:
arr = [64, 25, 12, 22, 11]
sorted_arr = quick_sort(arr)
print("Quick Sorted array:", sorted_arr)
```

# Bubble Sort

- Bubble Sort: A simple sorting algorithm that works by comparing adjacent pairs
of elements and swapping them if they are in the wrong order. It is easy to
implement, but has a running time of $O(n^2)$ in the worst case.

Español:

- Ordenación por burbujas: Un algoritmo de ordenación simple que funciona comparando pares adyacentes
adyacentes y los intercambia si están en el orden incorrecto. Es fácil de
implementar, pero tiene un tiempo de ejecución de $O(n^2)$ en el peor de los casos.

## Pseudocode/Pseudocódigo:

image.png

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Usage:
arr = [64, 25, 12, 22, 11]
bubble_sort(arr)
print("Bubble Sorted array:", arr)
```

# Radix Buckets

- Radix Sort: The radix sorting algorithm is a non-comparative sorting algorithm
that works by distributing elements into containers or buckets based on a
specific digit. The algorithm is executed several times, once for each digit,
moving elements from the buckets back to the original array on each iteration.
The time complexity of this algorithm is $O(nk)$ where k is the number of digits.

- Ordenación radix: El algoritmo de ordenación radix es un algoritmo de ordenación no comparativo
que funciona distribuyendo elementos en contenedores o cubos basados en un
un dígito específico. El algoritmo se ejecuta varias veces, una por cada dígito,
moviendo los elementos de los cubos de vuelta a la matriz original en cada iteración.
La complejidad temporal de este algoritmo es $O(nk)$ donde k es el número de dígitos.

## Pseucode/Pseudocódigo:

image.png

## Pseudocode (For the buckets)/Pseudocódigo (para los cubos/cubetas):

image.png

```python
def radix_bucket_sort(arr):
    # Find the maximum number to determine the number of digits
    max_num = max(arr)
    num_digits = len(str(max_num))

    # Create 10 buckets (0 to 9)
    buckets = [[] for _ in range(10)]

    # Perform counting sort for each digit from right to left
    for digit in range(num_digits):
        # Distribute elements into buckets
        for num in arr:
            digit_val = (num // 10 ** digit) % 10
            buckets[digit_val].append(num)

        # Collect elements from buckets
        arr = []
        for bucket in buckets:
            arr.extend(bucket)
            bucket.clear()

    return arr

# Example usage:
arr = [170, 45, 75, 90, 802, 24, 2, 66]
sorted_arr = radix_bucket_sort(arr)
print("Sorted array:", sorted_arr)
```

# Insertion Sort

- Insertion Sort: A simple sorting algorithm that works by inserting each element
into its correct place in a sorted list. It is easy to implement, but has a running
time of $O(n^2)$ in the worst case.

Español:

- Ordenación por inserción: Un algoritmo de ordenación simple que funciona insertando cada elemento
en su lugar correcto en una lista ordenada. Es fácil de implementar, pero tiene un tiempo de ejecución de $O(n^2)$ en el peor de los casos.
de $O(n^2)$ en el peor de los casos.

## Pseudocode/Pseudocódigo:

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Example usage:
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print("Sorted array:", arr)
```

# Counting Sort

- Counting Sort has a time complexity of $O(n + k)$, where '$n$' is the number of elements in the input array, and '$k$' is the range of input values. It is particularly efficient when the range of input values is not significantly larger than the number of elements in the array.

Español:

- Counting Sort tiene una complejidad temporal de $O(n + k)$, donde '$n$' es el número de elementos de la matriz de entrada y '$k$' es el rango de valores de entrada. Es especialmente eficiente cuando el rango de valores de entrada no es significativamente mayor que el número de elementos de la matriz.

## Pseudocode/Pseudocódigo:

image.png

```python
def counting_sort(arr):
    # Find the maximum and minimum values in the input array
    max_val = max(arr)
    min_val = min(arr)

    # Create a counting array to store the count of each element
    count_arr = [0] * (max_val - min_val + 1)

    # Count the occurrences of each element in the input array
    for num in arr:
        count_arr[num - min_val] += 1

    # Reconstruct the sorted array from the counting array
    sorted_arr = []
    for i in range(len(count_arr)):
        sorted_arr.extend([i + min_val] * count_arr[i])

    return sorted_arr

# Example usage:
arr = [4, 2, 2, 8, 3, 3, 1]
sorted_arr = counting_sort(arr)
print("Sorted Array:", sorted_arr)
```

# Dijkstra's Algorithm

- An algorithm for finding the shortest path in a graph with
non-negative edge weights. It works by visiting the vertices of the graph
incrementally, and updating the distances to adjacent vertices at each step. The
time complexity of Dijkstra's algorithm is $O(E + V log V)$, where E is the number
of edges and V is the number of vertices in the graph.

Español:

- Un algoritmo para encontrar el camino más corto en un grafo con
aristas no negativas. Funciona visitando los vértices del grafo
y actualizando las distancias a los vértices adyacentes en cada paso. La
La complejidad temporal del algoritmo de Dijkstra es $O(E + V log V)$, donde E es el número de aristas y V es el número de puntos.
de aristas y V es el número de vértices del grafo.

## Pseudocode/Pseudocódigo:

image.png

```python
import networkx as nx
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

def plot_graph(graph, node_colors, edge_colors):
    pos = nx.spring_layout(graph)

    nx.draw(graph, pos, with_labels=True, node_color=[node_colors[node] for node in graph.nodes], edge_color=[edge_colors[edge] for edge in graph.edges], node_size=500)
    plt.show()
    time.sleep(1)  # Adjust the sleep time to control animation speed
    clear_output(wait=True)

def dijkstra_visualization(graph, start_node):
    node_colors = {node: 'white' for node in graph}
    edge_colors = {edge: 'black' for edge in graph.edges}

    node_colors[start_node] = 'green'
    plot_graph(graph, node_colors, edge_colors)

    distances = {node: float('inf') for node in graph}
    distances[start_node] = 0

    unvisited_nodes = set(graph.nodes)
    current_node = start_node

    while unvisited_nodes:
        for neighbor in graph.neighbors(current_node):
            edge = (current_node, neighbor)
            new_distance = distances[current_node] + graph.edges[edge]['weight']
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                edge_colors[edge] = 'red'
                node_colors[neighbor] = 'green'
                plot_graph(graph, node_colors, edge_colors)
                edge_colors[edge] = 'black'
                node_colors[neighbor] = 'white'

        unvisited_nodes.remove(current_node)
        node_colors[current_node] = 'gray'
        plot_graph(graph, node_colors, edge_colors)

        min_distance = float('inf')
        for node in unvisited_nodes:
            if distances[node] < min_distance:
                min_distance = distances[node]
                current_node = node

    return distances

# Create a graph
G = nx.Graph()
G.add_edge('A', 'B', weight=2)
G.add_edge('A', 'C', weight=4)
G.add_edge('B', 'C', weight=1)
G.add_edge('B', 'D', weight=3)
G.add_edge('C', 'D', weight=5)

# Start Dijkstra's algorithm from node 'A'
start_node = 'A'
shortest_distances = dijkstra_visualization(G, start_node)
print("Shortest distances:", shortest_distances)
```

# Kruskal's Algorithm:

- Kruskal's Algorithm works by sorting the edges of the graph by weight and
adding the edges to the tree in order, as long as they do not cause cycles in the
tree. It uses a disjoint set (Union-Find) to determine whether an edge causes a
cycle or not. The time complexity of Kruskal's algorithm is O(E log E) in the best
case, where E is the number of edges in the graph. However, in cases where a
Union-Find implementation with a time complexity of O(log V) is used, the time
complexity is $O(E log V)$

Español:

- El Algoritmo de Kruskal funciona ordenando las aristas del grafo por peso y
y añadiendo las aristas al árbol en orden, siempre que no provoquen ciclos en el
árbol. Utiliza un conjunto disjunto (Union-Find) para determinar si una arista causa o no un ciclo.
ciclo o no. La complejidad temporal del algoritmo de Kruskal es O(E log E) en el mejor de los casos.
donde E es el número de aristas del grafo. Sin embargo, en los casos en que una
Union-Find con una complejidad temporal de O(log V), la complejidad temporal es O(E log V).
es $O(E log V)$.


## PseudoCode/Pseudocódigo:
image.png


```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph using NetworkX
G = nx.Graph()

# Add edges to the graph with weights
G.add_edge("A", "B", weight=4)
G.add_edge("A", "C", weight=2)
G.add_edge("B", "C", weight=1)
G.add_edge("B", "D", weight=5)
G.add_edge("C", "D", weight=8)

# Compute the MST using Kruskal's Algorithm
mst = nx.minimum_spanning_tree(G)

# Plot the original graph and the MST
plt.figure(figsize=(10, 5))

plt.subplot(121)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, font_size=10, node_color="lightblue")
labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Original Graph")

plt.subplot(122)
nx.draw(mst, pos, with_labels=True, node_size=500, font_size=10, node_color="lightgreen")
labels = nx.get_edge_attributes(mst, "weight")
nx.draw_networkx_edge_labels(mst, pos, edge_labels=labels)
plt.title("Minimum Spanning Tree")

plt.show()
```

# Bellman-Ford Algorithm

- The Bellman-Ford algorithm is an algorithm for finding the shortest path in a
graph with negative weights and/or negative cycles. It works by relaxing the
distances to the adjacent vertices several times until no better distances can be
found. The algorithm runs in $V-1$ iterations where V is the number of vertices in
the graph. The time complexity of the Bellman-Ford algorithm is $O(VE)$, where E
is the number of edges and $V$ is the number of vertices in the graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph using NetworkX
G = nx.DiGraph()  # Directed graph

# Add edges to the graph with weights
G.add_edge("A", "B", weight=2)
G.add_edge("A", "C", weight=4)
G.add_edge("B", "C", weight=1)
G.add_edge("B", "D", weight=7)
G.add_edge("C", "D", weight=3)
G.add_edge("D", "A", weight=6)

# Define the source node
source = "A"

# Compute shortest paths using Bellman-Ford Algorithm
shortest_paths = nx.single_source_bellman_ford_path_length(G, source)

# Plot the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, font_size=10, node_color="lightblue")
labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Directed Graph")

plt.show()
```

```python
# Display the shortest paths from the source node
print("Shortest Paths from Source Node:")
for target, distance in shortest_paths.items():
    print(f"{source} to {target}: {distance}")
```

# Warshall's Algorithm

- The Warshall algorithm is an algorithm for finding the shortest path between all
pairs of vertices in a graph. It works by updating the distances between the
vertices by relaxing the distances through an intermediate vertex. The time
complexity of the Warshall algorithm is $O(n^3)$ where n is the number of vertices
in the graph.

# Prim's Algorithm:

- An algorithm for finding a minimum spanning tree (MST) in a
connected and weighted graph. It works by visiting the vertices of the graph
incrementally, and adding the edges with the lowest weight that connect the
current subgraph to unvisited vertices. The time complexity of Prim's algorithm is
$O(E log V)$ in the best case, where E is the number of edges and V is the
number of vertices in the graph. However, in cases where the implementation is
based on an adjacency matrix, the time complexity is $O(V^2 + E)$.

# Depth-First Search (DFS):

The time complexity of a Depth-First Search $(DFS)$ algorithm depends on the data structure used to represent the graph and the specific implementation details. Here are the time complexities for common cases:

Adjacency List Representation (Graph as a Dictionary of Lists):

When a graph is represented as an adjacency list, the time complexity of DFS is typically $O(V + E)$, where V is the number of vertices $(nodes)$ and E is the number of edges in the graph.
Adjacency Matrix Representation:

When a graph is represented as an adjacency matrix, the time complexity of DFS is $O(V^2)$ since you need to check all possible edges.
In both cases, 'V' represents the number of vertices, and 'E' represents the number of edges in the graph.

Español:

# Búsqueda en profundidad (DFS):

La complejidad temporal de un algoritmo de Búsqueda por Profundidad $(DFS)$ depende de la estructura de datos utilizada para representar el grafo y de los detalles específicos de implementación. A continuación se muestran las complejidades temporales de los casos más comunes:

Representación de listas de adyacencia (grafo como diccionario de listas):

Cuando un grafo se representa como una lista de adyacencia, la complejidad temporal de DFS suele ser $O(V + E)$, donde V es el número de vértices $(nodos)$ y E es el número de aristas del grafo.
Representación de la matriz de adyacencia:

Cuando un grafo se representa como una matriz de adyacencia, la complejidad temporal de DFS es $O(V^2)$ ya que es necesario comprobar todas las aristas posibles.
En ambos casos, 'V' representa el número de vértices, y 'E' representa el número de aristas en el grafo.

## Pseudocode/Pseudocodigo:

image.png

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph using NetworkX
G = nx.Graph()

# Add edges to the graph
G.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (3, 7)])

# Perform DFS traversal starting from node 1
dfs_tree = nx.dfs_tree(G, source=1)

# Plot both the original graph and the DFS traversal tree side by side
plt.figure(figsize=(12, 5))

plt.subplot(121)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, font_size=10, node_color="lightblue")
plt.title("Original Graph")

plt.subplot(122)
pos = nx.spring_layout(dfs_tree)
nx.draw(dfs_tree, pos, with_labels=True, node_size=500, font_size=10, node_color="lightgreen")
plt.title("DFS Traversal Tree")

plt.show()
```

# Quantum Algorithms

Linear maps. Let us recall that a $\mathbb{C}$-linear map $T : \mathcal{H}^{(n)} \rightarrow \mathcal{H}^{(n)}$ (also called an operator) is determined by the $2^n$ images $t_j = T|j\rangle$, for:

\begin{align}
\mathbf{T} \left(\sum_{j} a_j|j\rangle\right) &= \sum_{j} a_jT(|j\rangle) \\
&= \sum_{j} a_jt_j
\end{align}
